# Анализ факторов, влияющих на уровень продаж компьютерных и консольных игр

  
### Описание проекта:

В данном проекте представлены исторические данные о продажах игр в разных регионах мира, оценки пользователей и экспертов, жанры и платформы за период с 1980 года по 2016 год включительно. 

Задача проекта - выявить определяющие успешность игр закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

### План проекта:

1. Предобработка данных
- Проверка внешнего вида данных и типа данных;
- Работа с пропусками;
- Работа с дубликатами.

2. Исследовательский анализ данных
- Как менялся выпуск игр по годам?
- Анализ динамики продаж игр наиболее популярных платформ;
- Какие платформы лидируют по продажам, растут или падают?
- Анализ распредления продаж в разбивке по потенциально прибыльным платформам;
- Как влияют отзывы пользователей и критиков на продажи популярных платформ?
- Анализ общего распределения жанров и их прибыльности.

3. Портрет пользователя относительно региона

Определить для пользователя каждого региона (NA, EU, JP):
- Самые популярные платформы (топ-5).
- Самые популярные жанры (топ-5).
- Влияет ли рейтинг ESRB на продажи в отдельном регионе?

4. Проверка статистических гипотез
- Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
- Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

### Описание данных

- Name — название игры
- Platform — платформа
- Year_of_Release — год выпуска
- Genre — жанр игры
- NA_sales — продажи в Северной Америке (миллионы проданных копий)
- EU_sales — продажи в Европе (миллионы проданных копий)
- JP_sales — продажи в Японии (миллионы проданных копий)
- Other_sales — продажи в других странах (миллионы проданных копий)
- Critic_Score — оценка критиков (максимум 100)
- User_Score — оценка пользователей (максимум 10)
- Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats as st
```

```python
data = pd.read_csv('games_dataset_for_2_project.csv')
display(data.head())
```

# Раздел 1: Предобработка данных

```python
lower_columns = []
for column in data.columns:
    lower_columns.append(column.lower())
data.columns = lower_columns
data.head()
```

<div class="alert alert-info">
    Далее необходимо поработать с пропусками данных и типами данных, поэтому на первом этабе необходимо посмотреть общую информацию о данных.

```python
display(data.info())
data.isna().sum()
```

<div class="alert alert-info">
    Строки с пропусками в стобцах name и genre не значительны (всего 2), поэтому можно удалить эти строки. Интересно проверить много ли данных, где имеется пропуск в трех столбцах одновременно: critic_score, user_score, rating. Кроме того есть предположение, что большая доля пропусков в данных до 2000-х годов, поэтому интересно посмотреть список уникальных значений года в строках с пропусками.

```python
data.dropna(subset=['name', 'genre'], inplace=True)
display(data.shape)
data_for_analysis_pass = data.loc[(data['critic_score'].isna() == True) & \
                                  (data['user_score'].isna() == True) & (data['rating'].isna() == True)]
display(data_for_analysis_pass)
display(data_for_analysis_pass['year_of_release'].unique())
data_for_analysis_pass.loc[data['year_of_release'] <= 2000]
```

<div class="alert alert-info">Следующий этап - обработка значения tbd в столбце user_score.

```python
display(data['user_score'].unique())
display((data[data['user_score'] == 'tbd']).head())
data['user_score'] = data['user_score'].replace(['tbd'], np.NaN)
display(data['user_score'].unique())
data['user_score'] = data['user_score'].astype('float')
data.info()
```

<div class="alert alert-info"> 
    
В столбце year_of_release имеется 269 пропусков. Появление пропусков возможно в той ситуации, когда релиз игры еще не произошел. Но при этом имеются продажи - это может быть предзаказ игры. Заменить пропуски в этом столбце не представляется возможным, удалить эти строки будет не логичным, так как в этих строках есть другие важные данные.

Более чем в половине данных имеется пропуск в столбце critic_score, и в 40% данных имеются пропуски в столбцах user_score и rating. Данные пропуски не возможно заменить на среднее или медианное значение, а удалить половину DataFrame будет не правильно. При этом пропуски встречаются на всем периоде выпуска игр: как до 2000 года так и после. Можно предположить, что пропуски в значениях оценки игр более раннего периода появились, так как такая система оценки не использовалось в то время, например в 80-е годы 20 века. 

Отдельно рассмотрим значение tbd в столбце user_score. TBD рассшифровывается как To Be Determined, то есть "будет определено". Получается что на данный момент оценка пользователей по этим играм не определена, в целом это значение схоже со значением nan, поэтому приведем столбец к порядку: оставим численные значения и пропущенные (nan), это действие также позволит привести столбец к типу float.

```python
data['sum_sales'] = data['na_sales'] + data['eu_sales'] + data['jp_sales'] + data['other_sales']
data.head()
```

<div class="alert alert-info">
Добавлен столбец с общей суммой продаж во всех регионах.

# Раздел 2: Исследовательский анализ данных

```python
data.pivot_table(index='year_of_release', values='name', aggfunc='count') \
    .rename(columns = {'year_of_release' : 'year', 'name' : 'released games',}) \
    .plot(kind='line', figsize=(20,10),  style = 'o-', grid=True)
plt.title('Количество выпущенных игр по годам'+ "\n")
plt.xlabel('Год выпуска')
plt.ylabel('кол-во')
```

<div class="alert alert-info">
Согласно графику мы видим, что в период с 1980 по 1993 год число выпущенных игр близко к нулю, поэтому данные за этот период не важны для анализа. В дальнейшем число выпущенных игр стремительно растет, пик выпуска произошел в 2008-2009 годах, около 1400 игр в год. Видимо в этот период произошло перенасыщение рынка, поэтому с 2009 года происходит спад выпуска игр. 

```python
the_most_profitable_platforms = data.pivot_table(index='platform', values='sum_sales', aggfunc='sum') \
     .sort_values(by='sum_sales', ascending=False) \
     .head(10) \
     .reset_index()
display(the_most_profitable_platforms)

platforms = data.query('platform in @the_most_profitable_platforms["platform"]')
display(platforms.head())

display(platforms['platform'].unique())

platforms = platforms.pivot_table(index=['year_of_release', 'platform'], values='sum_sales', aggfunc='sum' )
platforms = platforms.reset_index()
platforms.head()
```

<div class="alert alert-info">
Подготовлены данные по 10 игровым платформам с наибольшими объемами продаж по всему миру.

```python
groups = platforms.groupby('platform')
fig, ax = plt.subplots()
for name, group in groups:
    ax.plot(group['year_of_release'], group['sum_sales'], marker="o", linestyle="-", label=name)
fig.set_figwidth(15)
plt.title('Динамика продаж по платформам'+ "\n")
plt.xlabel('Год выпуска')
plt.ylabel('Сумма продаж')
plt.legend()
plt.grid(True)
plt.show()
```

<div class="alert alert-info">

Так как PS3 и X360 - платформы прошлого поколения. #### Анализ влияния отзывов пользователей на продажи самой популярной платформы 'PS3'.

```python
data_since_2014 = data[data['year_of_release'] >= 2014]
data_of_ps3 = data_since_2014.loc[data['platform'] == 'PS3']
display(data_of_ps3.head())

data_of_ps3.plot(x='critic_score', y='sum_sales', kind='scatter', grid=True)
plt.title('Взаимосвязь оценок критиков и  продаж игр')

data_of_ps3.plot(x='user_score', y='sum_sales', kind='scatter', grid=True)
plt.title('Взаимосвязь оценок пользователей и  продаж игр')

display(data_of_ps3['sum_sales'].corr(data_of_ps3['critic_score']))
data_of_ps3['sum_sales'].corr(data_of_ps3['user_score'])
```

<div class="alert alert-info">
Согласно данным графика можно сделать следующий вывод: оценки критиков демонстрируют прямую, средне-слабую корреляцию с "цифрами" продаж. Оценки пользователей корреляции с продажами не имеют. На графике видно, что не смотря на невысокие оценки пользователей, есть случаи больших продаж.

```python
for platform in potential_platforms:
    print(platform)
    data_of_platform = data_since_2014.loc[data['platform'] == platform]
    print(data_of_platform['critic_score'].corr(data_of_platform['sum_sales']))
    print(data_of_platform['user_score'].corr(data_of_platform['sum_sales']))
```

<div class="alert alert-info">
По другим платформам между продажами и оценками критиков имеется прямопропорциональная слабая и средня зависимость. Между оценками пользователей и продажами зависимости нет.

```python
genre = data_since_2014.pivot_table(index='genre', values='sum_sales', aggfunc=['count','mean','sum'])
genre.columns = ['count_sales', 'mean_sales', 'sum_sales']
genre = genre.sort_values(by='count_sales', ascending=False)
genre
```

<div class="alert alert-info">
Согласно расчетам, наиболее популярным является жанр игр - Action, однако более ценным является жанр игр - Shooter показали результаты, в среднем игра продается в размре 1,3 млн копий. Средняя продажа сильно отличается от данных по другим жанрам, можно сказать, что это самый прибыльный жанр игр.

# Раздел 3: Портрет пользователя относительно региона

<div class="alert alert-info">
Сравним популярность игровых платформ в разных регионах.

```python
regions = ['na_sales', 'eu_sales', 'jp_sales']
for region in regions:
    print(region)
    platforms_in_region = data_since_2014.pivot_table(index='platform', values=region, aggfunc=['sum'])
    platforms_in_region.columns=['sum_sales']
    platforms_in_region = platforms_in_region.reset_index().sort_values(by='sum_sales', ascending=False)
    plt.pie(platforms_in_region['sum_sales'], labels=platforms_in_region['platform'], autopct='%1.1f%%', radius=1.5)
    plt.show()
```

<div class="alert alert-info">
    
Для жителей Северной Америки наиболее популярными игровыми платформами являются PS4, Xone, X360. 35% всех продаж за период с 2014 по 2016 год были на платформе PS4.

Для жителей Европы также наиболее популярной 
    платформой является  PS4. На этой платформе была осуществлена почти половина всех продаж.

Для жителей Японии платформа PS4 не так популярна. На ней было продано около 16% от общей суммы продаж. Лидером продаж является платформа 3DS - 47% продаж. На японском рынке популярны только японские платформы, продажи на платформе X360 близки к нулю.

<div class="alert alert-info"> Сравним популярность жанров игр по регионам.

```python
for region in regions:
    print(region)
    genres_in_regions = data_since_2014.pivot_table(index='genre', values=region, aggfunc=['sum'])
    genres_in_regions.columns=['sum_sales']
    genres_in_regions = genres_in_regions.reset_index().sort_values(by='sum_sales', ascending=False)
    plt.pie(genres_in_regions['sum_sales'], labels=genres_in_regions['genre'], autopct='%1.1f%%', radius=1.5)
    plt.show()
```

<div class="alert alert-info">
    
Согласно данным диаграмм предпочтения пользователей из Северной Америки и Европы в выборе игр по жанрам схожи: в Топ игр входят игры жанра Action, Sports, Shooter. В целом по этим регионам от 25 до 28% всех продаж игр приходятся на игры жанра Action и Shooter,.

Пользователи Японии имеют другие предпочтения. Теть всех продаж относится на игры жанра Role-Playing. Продажи игр жанра Action близки к объему продаж жанра Role-Playing. Спрос на игры этих жанров сильно выделяется, 60% рынка приходятся на игры этих жанров. Далее продажи игр по остальным жанрам не превышают 7%.

Можно сказать, что в Северной Америке и Европе имеется больший разброс в предпочтениях жанра игр, в Японии выбор жанра более узкий, в основном ограничен 3 жанрами: Role-Playing и Action.

<div class="alert alert-info">
Проведем анализ популярности игр по их рейтингу в разных регионах. В столбце 'rating' имеются значения NaN, чтобы учитывать продажи игр, в том числе и без рейтинга, заменим пропуски значений на 'not defined'.

```python
display(data_since_2014['rating'].value_counts(dropna=False))
data_since_2014['rating'] = data_since_2014['rating'].fillna('not defined')
for region in regions:
    print(region)
    rating_in_regions = data_since_2014.pivot_table(index='rating', values=region, aggfunc='sum')
    rating_in_regions.columns=['sum_sales']
    rating_in_regions = rating_in_regions.reset_index().sort_values(by='sum_sales', ascending=False)
    plt.pie(rating_in_regions['sum_sales'], labels= rating_in_regions['rating'], autopct='%1.1f%%')
    plt.show()
```

<div class="alert alert-info">
Продажи игр рейтинга М (от 17 лет) в Европе и Северной Америке больше, чем в Японии, в свою очередь в Японии больше продаж игр рейтинга Т (то 13 лет). Продажи игр рейтинга Е10+ примерно одинаковы во всех регионах. В Японии более 60% проданных игр не имеют рейтинга ESRB. Это связано с тем что в Японии собственная система оценки игр - CERO.

# Раздел 4: Проверка гипотез

<div class="alert alert-info"> Необходимо использовать метод для проверки гипотезы о равенстве среднего двух независимых генеральных совокупностей scipy.stats.ttest_ind.

<div class="alert alert-info">
Нулевая гипотеза - Средние пользовательские рейтинги платформ Xbox One и PC одинаковые; 

Тогда альтернативная гипотеза будет звучать так: Средние пользовательские рейтинги платформ Xbox One и PC различаются;

```python
data_of_x360 = data_since_2014.query('platform == "X360"').dropna(subset=['user_score'])
display(data_of_x360.head())
display(data_of_x360.info())

data_of_pc = data_since_2014.query('platform == "PC"').dropna(subset=['user_score'])
display(data_of_pc.head())
display(data_of_x360.info())
```

```python
alpha = 0.05
results = st.ttest_ind(data_of_x360['user_score'], data_of_pc['user_score'])
print(results)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else: 
    print('Не получилось отвергнуть нулевую гипотезу')
```

```python
fig, ax = plt.subplots()
data_of_x360.pivot_table(index='user_score', values='name', aggfunc='count').plot(figsize=(15,5), grid=True, ax=ax)
data_of_pc.pivot_table(index='user_score', values='name', aggfunc='count').plot(figsize=(15,5), grid=True, ax=ax)
plt.title('Оценки пользователей игр разных платформ')
plt.legend(['XBOX360', 'PC'])
plt.show()
```

<div class="alert alert-info">
Pvalue составило 0.030983578725445982 (3%) - это слишком маленькая вероятность, чтобы делать вывод о сходстве среднего значения оценки пользователей платформы Xbox One и PC. Графики также демонстрируют различие пользовательских оценок.

<div class="alert alert-info">
Нулевая гипотеза - Средние пользовательские рейтинги жанров Action и Sports одинаковые; 

Тогда альтернативная гипотеза будет звучать так: Средние пользовательские рейтинги жанров Action и Sports разные;

```python
data_of_action = data_since_2014.query('genre == "Action"').dropna(subset=['user_score'])
display(data_of_action.head())
display(data_of_action.info())

data_of_sports = data_since_2014.query('genre == "Sports"').dropna(subset=['user_score'])
display(data_of_sports.head())
display(data_of_sports.info())
```

```python
results = st.ttest_ind(data_of_action['user_score'], data_of_sports['user_score'])
print(results)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else: 
    print('Не получилось отвергнуть нулевую гипотезу')
```

```python
fig, ax = plt.subplots()
data_of_action.pivot_table(index='user_score', values='name', aggfunc='count').plot(figsize=(15,5), grid=True, ax=ax)
data_of_sports.pivot_table(index='user_score', values='name', aggfunc='count').plot(figsize=(15,5), grid=True, ax=ax)
plt.title('Оценки пользователей игр разных жанров')
plt.legend(['Action', 'Sports'])
plt.show()
```

<div class="alert alert-info">
Pvalue составило 2.8711147985105864e-196  - это слишком маленькая вероятность, чтобы делать вывод о сходстве среднего значения оценки пользователей игр жанра Action и Sports. Согласно графику видно, что пользовательские оценки игр жанра Action в большинстве случаев выше, чем оценки игр жанра Sports.
